#  Question 6

Create the below pattern using nested for loop in Python.
*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*

```python
for i in range(10):
    if i<5:
        for j in range(i+1):
            print('* ',end='')
        print()
    else:
        for k in range(i,9):
            print('* ',end='')
        print()
```
